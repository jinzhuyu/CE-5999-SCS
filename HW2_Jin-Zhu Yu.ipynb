## 3    Population dynamics
### 3.2   Is the assumption on the growth rate realistic?
 Yes, that is a valide assumption. In the absence of competitors, predators, or environmental constraints, such as food, the growth rate increases rapidly with the increase of the total population. The totoal population increases exponentially, which matchs the solution of the equation: $r=C\times exp[(b_r-d_r)t]$ where C is a constant to be determined by the initial condition.

### 4.10 Write a function called pred_prey

```python

def pred_prey(y,t): 
# return the devirative of y_dot(t) at t given y(t) # nonindent shortcut: shift+tab
# inputs: 
# y=[r,w], array_like 
# t=time, array_like 
# output: 
# y_dot=[r_dot,w_dot]
   
    b_r=0.05
    d_r=0.01
    b_w=0.1
    d_w=0.2
    k_r=0.001
    e_w=0.002
    
    r_dot=(b_r - d_r - k_r*y[1])*y[0]    
    w_dot=(b_w - d_w + e_w*y[0])*y[1]
    
    y_dot=[r_dot,w_dot]   
    
    return y_dot
```

### 4.13  Derive the solution the coupled ODE equations
The predator-prey equations, also known as the [Lotka–Volterra equations](https://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equations).

Generali idea: eliminate $dt$ from the two equations and then apply the seperation method.


Let $A=b_r-d_r$ and $B=b_w-d_w$

$\begin{array}{l}
\left\{ \begin{array}{l}
\frac{{dr}}{{dt}} = \left( {A - kw} \right)r\\
\frac{{dw}}{{dt}} = \left( {B + er} \right)w
\end{array} \right.\\
\\
\left\{ \begin{array}{l}
\frac{{dr}}{{dt}}\frac{1}{{rw}} = \left( {A - kw} \right)r\frac{1}{{rw}} = \frac{A}{w} - k\\
\frac{{dw}}{{dt}}\frac{1}{{rw}} = \left( {B + er} \right)w\frac{1}{{rw}} = \frac{B}{r} + e
\end{array} \right.\\
\\
\left\{ \begin{array}{l}
\frac{{dr}}{{dt}}\frac{1}{{rw}}\frac{{dw}}{{dt}} = \frac{{dw}}{{dt}}\left( {\frac{A}{w} - k} \right)\\
\frac{{dw}}{{dt}}\frac{1}{{rw}}\frac{{dr}}{{dt}} = \frac{{dr}}{{dt}}\left( {\frac{B}{r} + e} \right)
\end{array} \right.\\
\\
\frac{{dw}}{{dt}}\left( {\frac{A}{w} - k} \right) - \frac{{dr}}{{dt}}\left( {\frac{B}{r} + e} \right) = 0\begin{array}{*{20}{c}}
{}&{\left( a \right)}
\end{array}\\
\\
{\textrm{Suppose the solution is some function }} f\left( {r,w} \right)\\
\frac{{df}}{{dt}} = \frac{{\partial f}}{{\partial w}}\frac{{dw}}{{dt}} + \frac{{\partial f}}{{\partial r}}\frac{{dr}}{{dt}} = 0\begin{array}{*{20}{c}}
{\begin{array}{*{20}{c}}
{}&{}
\end{array}}&{\left( b \right)}
\end{array}\\
\\
{\textrm{Comparing }}(a){\textrm{ and }}(b){\textrm{ yields:}}\\
\left\{ \begin{array}{l}
\frac{{\partial f}}{{\partial w}} = \frac{A}{w} - k\\
\frac{{\partial f}}{{\partial r}} =  - \frac{B}{r} - e
\end{array} \right.\\
\\
f\left( {r,w} \right) = \int {\left( {\frac{A}{w} - k} \right)} \partial w = \int {\left( { - \frac{B}{r} - e} \right)} \partial r\\
\\
u(r) = \int {\left( { - \frac{B}{r} - e} \right)} \partial r =  - B\ln (r) - er + {C_1}\\
v(w) = \int {\left( {\frac{A}{w} - k} \right)} \partial w = A\ln (w) - kw + {C_2}\\
\\
f\left( {r,w} \right) = A\ln (w) - kw + u(r) =  - B\ln (r) - er + v(w)\\
\\
\textrm{Finally, we have:}\\
